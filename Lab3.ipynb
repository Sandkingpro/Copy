```python
import pandas as pd

heart_data = pd.read_csv('/content/heart.csv')
# show the first 5 lines of the dataframe
heart_data.head()
test_train_ratio=["0.4:0.6","0.3:0.7","0.2:0.8","0.1:0.9"]
accs=[]
recalls=[]
f1s=[]
```

```python
heart_data_x = heart_data.drop('HeartDisease', axis=1)
heart_data_y = heart_data['HeartDisease']
```

```python
heart_data_x_encoded = pd.get_dummies(heart_data_x, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(heart_data_x_encoded, heart_data_y, test_size=0.4)
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(heart_data_x_encoded, heart_data_y, test_size=0.3)
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(heart_data_x_encoded, heart_data_y, test_size=0.2)
```

```python
X_train3, X_test3, y_train3, y_test3 = train_test_split(heart_data_x_encoded, heart_data_y, test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,recall_score,f1_score
dtree = DecisionTreeClassifier(max_depth=4, criterion = 'gini')
dtree.fit(heart_data_x_encoded, heart_data_y)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
def print_tree():
  fig = plt.figure(figsize=((25,20)))
  plot_tree(dtree,
              feature_names = heart_data_x_encoded.columns,
              class_names=['no heart disease', 'heart disease'], 
              impurity=False,
              proportion=True,
              filled=True)
  fig.savefig('test.png')
print_tree()
```

```python

dtree = DecisionTreeClassifier(max_depth=4, criterion = 'gini')
dtree.fit(X_train, y_train)
predicitions = dtree.predict(X_test)
acc = accuracy_score(y_test, predicitions)
recall=recall_score(y_test, predicitions)
f1=f1_score(y_test,predicitions)
print_tree()
acc,recall,f1
accs.append(acc)
recalls.append(recall)
f1s.append(f1)
```

```python
dtree = DecisionTreeClassifier(max_depth=4, criterion = 'gini')
dtree.fit(X_train1, y_train1)
predicitions = dtree.predict(X_test1)
acc = accuracy_score(y_test1, predicitions)
recall=recall_score(y_test1, predicitions)
f1=f1_score(y_test1,predicitions)
print_tree()
acc,recall,f1
accs.append(acc)
recalls.append(recall)
f1s.append(f1)
```

```python
dtree = DecisionTreeClassifier(max_depth=4, criterion = 'gini')
dtree.fit(X_train2, y_train2)
predicitions = dtree.predict(X_test2)
acc = accuracy_score(y_test2, predicitions)
recall=recall_score(y_test2, predicitions)
f1=f1_score(y_test2,predicitions)
print_tree()
acc,recall,f1
accs.append(acc)
recalls.append(recall)
f1s.append(f1)
```

```python
dtree = DecisionTreeClassifier(max_depth=4, criterion = 'gini')
dtree.fit(X_train3, y_train3)
predicitions = dtree.predict(X_test3)
acc = accuracy_score(y_test3, predicitions)
recall=recall_score(y_test3, predicitions)
f1=f1_score(y_test3,predicitions)
print_tree()
acc,recall,f1
accs.append(acc)
recalls.append(recall)
f1s.append(f1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,5))
sns.barplot(x = test_train_ratio, y = accs, palette = 'gnuplot')
plt.xlabel('ratio', size = 10)
plt.xticks(rotation=45)
plt.ylabel('accuracy_score', size = 10)
plt.title('', color = 'green', size = 10)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x = test_train_ratio, y = recalls, palette = 'gnuplot')
plt.xlabel('ratio', size = 10)
plt.xticks(rotation=45)
plt.ylabel('recall_score', size = 10)
plt.title('', color = 'green', size = 10)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x = test_train_ratio, y = f1s, palette = 'gnuplot')
plt.xlabel('ratio', size = 10)
plt.xticks(rotation=45)
plt.ylabel('f1_score', size = 10)
plt.title('', color = 'green', size = 10)
plt.show()
```
